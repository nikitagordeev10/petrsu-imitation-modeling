```python
from numpy import array
from matplotlib.pyplot import figure, show
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Чтение данных из файла
def read_points(filename):
    points = []
    with open(filename, 'r', encoding='utf-8-sig') as file: 
        for line in file:
            # Заменяем запятые на точки, если они есть
            line = line.replace(',', '.')
            # Разделяем строку на числа и преобразуем их в float
            x, y, z = map(float, line.strip().split())
            points.append((x, y, z))
    return array(points)

# Построение геликоида
def plot_helix(points):
    fig = figure()
    ax = fig.add_subplot(111, projection='3d')

    # Разделяем координаты
    x = points[:, 0]
    y = points[:, 1]
    z = points[:, 2]

    # Построение точек
    ax.scatter(x, y, z, c='b', marker='o')

    # Настройка осей
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Показать график
    show()
```

```python
filename = "D:/petrsu/imitation_modeling/lib/helix_points.txt"   
points = read_points(filename)
plot_helix(points)
```
